Бачимо що є деякі missing values, але дублікатів чи інших пошкоджених даних не спостерігається. Перевіримо також кореляції про всяк випадок

```python
corr_mx_train = train_data.corr()
mask_train = np.array(corr_mx_train)
mask_train[np.tril_indices_from(mask_train)] = False
fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
sns.heatmap(corr_mx_train, cmap = "Reds", mask = mask_train, square = True)
```

```python
corr_mx_test = test_data.corr()
mask_test = np.array(corr_mx_test)
mask_test[np.tril_indices_from(mask_test)] = False
fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
sns.heatmap(corr_mx_test, cmap = "Reds", mask = mask_test, square = True)
```

Завеликих кореляцій не спостерігається. Але маємо декілька фіч, що потребують обробки. Наприклад імена, які унікальні для кожного запису ніяк не допоможуть нам при кластеризації. Також забагато унікальних значень у колонці Ticket, і при кодуванні даної категоріальної колонки one hot encoding-ом з'явиться забагато додаткових стовпців і модель буде схильна до перенавчання. Схожа проблема і з колонкою Cabin, але в ній ще й критично великий відсоток missing values. Їхнє заповнення, наприклад, медіаною може призвести до того, що модель буде занадто відхилена від реальних значень. Найпростішим і найшвидшим рішенням буде пожертвувати даними колонками.

```python
train_data = train_data.drop("Name", axis=1)
train_data = train_data.drop("Cabin", axis=1)
train_data = train_data.drop("Ticket", axis=1)

test_data = test_data.drop("Name", axis=1)
test_data = test_data.drop("Cabin", axis=1)
test_data = test_data.drop("Ticket", axis=1)
```

Заповнимо ж медіанами missing values у колонках, в яких їхня кількість відносно невелика

```python
train_data["Embarked"].fillna(value="S", limit=2, inplace=True)
train_data["Age"].fillna(train_data["Age"].mean(), inplace=True)

test_data["Age"].fillna(test_data["Age"].mean(), inplace=True)
test_data["Fare"].fillna(test_data["Fare"].mean(), inplace=True)
```

Тепер закодуємо категоріальні ознаки. Наприклад Embarked за допомогою one hot encoding та стать за допомогою label encoding-у

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

embarked = train_data[["Embarked"]]
ohe = OneHotEncoder()
encode = LabelEncoder()

train_data["Sex"] = encode.fit_transform(train_data["Sex"])
embarked_ohe = ohe.fit_transform(embarked).toarray()

embarked_ohe_df = pd.DataFrame(embarked_ohe, columns=ohe.get_feature_names_out(["Embarked"]))
train_data = train_data.join(embarked_ohe_df)

train_data.drop(["Embarked"], axis=1, inplace=True)

train_data.sample(5)
```

```python
embarked = test_data[["Embarked"]]

test_data["Sex"] = encode.fit_transform(test_data["Sex"])
embarked_ohe = ohe.fit_transform(embarked).toarray()

embarked_ohe_df = pd.DataFrame(embarked_ohe, columns=ohe.get_feature_names_out(["Embarked"]))
test_data = test_data.join(embarked_ohe_df)

test_data.drop(["Embarked"], axis=1, inplace=True)

test_data.sample(5)
```

Імпортуємо все необхідне для наступної кластеризації

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

Підготуємо до неї дані

```python
X_train = train_data.iloc[:, 1:]
X_test = test_data.iloc[:, 1:]

scaler = StandardScaler()
X_scaled_train = scaler.fit_transform(X_train)
X_scaled_test = scaler.fit_transform(X_test)
```

Проведемо кластеризацію трьома різними алгоритмами: DBSCAN, Agglomerative та K-Means Clustering. Спробуємо для них різні параметри, епсілон для DBSCAN та n_clusters для інших

```python
dbscan = DBSCAN(eps=3)
dbscan.fit(X_scaled_train)
labels = dbscan.labels_
train_data["Cluster_dbscan_3"] = labels

dbscan = DBSCAN(eps=2)
dbscan.fit(X_scaled_train)
labels = dbscan.labels_
train_data["Cluster_dbscan_2"] = labels

agg = AgglomerativeClustering(n_clusters=3)
agg.fit(X_scaled_train)
labels = agg.labels_
train_data["Cluster_agg_3"] = labels

agg = AgglomerativeClustering(n_clusters=5)
agg.fit(X_scaled_train)
labels = agg.labels_
train_data["Cluster_agg_5"] = labels

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled_train)
labels = kmeans.labels_
train_data["Cluster_kmeans_3"] = labels

kmeans = KMeans(n_clusters=16, random_state=42)
kmeans.fit(X_scaled_train)
labels = kmeans.labels_ 
train_data["Cluster_kmeans_16"] = labels
```

Оцінимо, яке відображення найкраще для хоча б приблизної наглядності кластерів даних

```python
sns.pairplot(data=train_data, vars=X_train.columns, hue="Cluster_dbscan_3", palette="bright")
```

16 кластерів це забагато для сприйняття (що, в теорії, було очевидно), отже і параметр eps=2 для DBSCAN не найкращий. Виконаємо ті ж самі дії для тестового датасету з більш-менш оптимальними для візуальної оцінки параметрами

```python
dbscan = DBSCAN(eps=3)
dbscan.fit(X_scaled_test)
labels = dbscan.labels_
test_data["Cluster_dbscan_3"] = labels

agg = AgglomerativeClustering(n_clusters=3)
agg.fit(X_scaled_test)
labels = agg.labels_
test_data["Cluster_agg_3"] = labels

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled_test)
labels = kmeans.labels_
test_data["Cluster_kmeans_3"] = labels
```
